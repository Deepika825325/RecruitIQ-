```python
import json

with open("../data/raw/sample_candidates.json") as f:
    candidates = json.load(f)

ground_truth = {
    "CAND_0000031": "strong_fit",
    "CAND_0000024": "reject",
    "CAND_0000015": "borderline_low",
    "CAND_0000010": "borderline_mid",
}

for c in candidates:
    if c["candidate_id"] in ground_truth:
        print(c["candidate_id"], "->", ground_truth[c["candidate_id"]], "|", c["profile"]["current_title"])
```
